```python
def fizzbuzz(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

# Call the function for numbers 1 to 20
fizzbuzz(20)
```

```python

todo_list = []

def add_task(task):
    todo_list.append(task)

def show_tasks():
    if not todo_list:
        print("Your to-do list is empty.")
    else:
        print("To-do list:")
        for index, task in enumerate(todo_list, 1):
            print(f"{index}. {task}")


add_task("Finish my food")
add_task("Buy groceries")
add_task("Do dishes")


show_tasks()
```

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
temperatures_in_celsius = [22, 46, 51, 76]

for temp in temperatures_in_celsius:
    fahrenheit = celsius_to_fahrenheit(temp)
    print(f"{temp}ºC = {fahrenheit:.2f}ºF")
```
